```python
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import Point
```

```python
from fpsnippets import geo_to_neighborhood
```

```python
restaurants = pd.read_csv("restaurants.csv")
restaurants.head()
```

## Bar Graph

After obtaining the data, I used the latitude and longitude that correlated with each restaurant to find which neighborhood it was loacated in. I then counted each restaurant in each neighborhood and plotted the results using a bar chart.

```python
# use geo_to_neighborhood function to convert lat and long 
restaurants['Neighborhood'] = restaurants.apply(lambda row: geo_to_neighborhood(row['Latitude'], row['Longitude']), axis=1)

# Count # of restaurants in each neighborhood
neighborhood_counts = restaurants['Neighborhood'].value_counts()

# Create bar chart
plt.figure(figsize=(15, 8))
neighborhood_counts.plot(kind='bar', color= "#7698FF")
plt.title('Restaurants in each neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Restaurants')
plt.show()
```

## Pie Chart

I next created a pie chart to gain even more clarification what neighborhoods stand out and see their precentages compared to other neighborhoods

```python
#Create pie chart
neighborhood_counts.name = " "
plt.figure(figsize=(15, 8)) 
colors = ['#F6A8B6', '#A0BAFF', 'lightblue', '#FFCB86', '#C1BAFF', 'skyblue', '#A4F4A3', '#3ED5F6', '#A7B2DB', '#748FF0']
neighborhood_counts.plot(kind='pie', autopct='%1.1f%%', colors = colors)
plt.title('Restaurant Counts by Neighborhood')
plt.show()
```

## Top 10 Neighborhoods

```python
#Sort neighborhoods from greatest # to lowest
sorted_neighborhood_counts = neighborhood_counts.sort_values(ascending=False)

# Filter top 10 neighborhoods
top_10_neighborhoods = sorted_neighborhood_counts.head(10)
colors = ['#F6A8B6', '#A0BAFF', 'lightblue', '#FFCB86', '#C1BAFF', 'skyblue', '#A4F4A3', '#3ED5F6', '#A7B2DB', '#748FF0']
# Create pie chart
plt.figure(figsize=(15, 8)) 
neighborhood_counts.name = " "
top_10_neighborhoods.plot(kind='pie', autopct='%1.1f%%', colors = colors)
plt.title('Top 10 Neighborhoods with the Most Restaurants')
plt.show()
```
