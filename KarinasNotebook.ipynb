## Personal Notebook: Demographic Submetric

#### In order to asses what neighborhood has the most fun, from the perspective of young adult college students we wanted to look at what area in Pittsburgh has the most single-young adults to fit our metric of what neighborhood is the most "fun". For this I found a demographic data set that breaks up the estimated households by family and non-family. Although I couldn't find a data set that represented the population by age and neighborhood, I feel as though abstracting from this data to find the single member-nonfamily households to determine where the most single-young adults are in Pittsburgh.

```python
import pandas as pd

data = pd.read_csv("households.csv")
data['Single_household_total_ratio'] = data['Estimate; Nonfamily households: - Householder living alone'] / data['Estimate; Total:']

data.head(10)
```

```python
selected_columns = ['Neighborhood','Single_household_total_ratio']
data_subset = data[selected_columns]

print(data_subset)
```

```python
specified_neighborhoods = ['Shadyside', 'Central Oakland', 'Central Business District', 'Strip District','Carrick','South Side Flats','South Oakland','North Oakland','Mount Washington']

specified_data = data_subset[data['Neighborhood'].isin(specified_neighborhoods)]
print(specified_data) #
```

```python
import matplotlib.pyplot as plt #Importing matplot for access to Bar graph

SdPlot = specified_data.plot(kind='bar', x = 'Neighborhood',y = 'Single_household_total_ratio') #creatig a plot with my specified data
```

### Conclusion

Based on my data, it appears that the neighborhood that has the most fun is the Strip District. Out of all household types, the Strip District has the most single non-family member households. I did not include single parent family households due to the likkelihood that this household would be out of the 18-25 year old age range that was intended for this data or as our target demographic. Things makes sense due to the trendiness of the neighborhood and the liklihood of young single people moving there after graduating.
